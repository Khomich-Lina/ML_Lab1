```python
import numpy as np
from numpy import genfromtxt
```

```python
my_data = genfromtxt('lab_1_train.csv',delimiter=',')
```

```python
x = my_data[:,1]
y = my_data[:,2]

x = x[~np.isnan(x)]
y = y[~np.isnan(y)]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('train.csv')
plt.show()
```

```python
def Mean_Squared_Error(teta_zero, teta_one, x,y):
    return 1/len(x)*sum(((teta_zero + teta_one*x)-y)**2)
```

```python
import random
```

```python
def gradient_descent(x, y, learning_rate, teta_tolerance, iteration_tolerance ):
    teta_zero = random.uniform(0,1)
    teta_one = random.uniform(0,1)
    
    CF_start = Mean_Squared_Error(teta_zero, teta_one,x,y)
   
     
    for i in range(iteration_tolerance):
        teta_zero_temp = teta_zero - learning_rate/len(x) * np.sum((teta_zero+ teta_one*x)-y)
        teta_one_temp = teta_one - learning_rate/len(x) * np.sum(((teta_zero+ teta_one*x)-y)*x)
        
        cost_function = Mean_Squared_Error(teta_zero_temp, teta_one_temp,x,y)
        
        output= F"[{i}]: teta_zero - {teta_zero_temp}; teta_one - {teta_one_temp}; Cost Function - {cost_function}"
        print(output)
        
        
        
        if abs (teta_zero_temp - teta_zero)<teta_tolerance and abs(teta_one_temp - teta_one)<teta_tolerance:
            return teta_zero_temp, teta_one_temp
        teta_one = teta_one_temp
        teta_zero = teta_zero_temp
        
    return teta_zero, teta_one
        
    
```

```python
x = x
y = y

teta_zero, teta_one = gradient_descent(x, y, learning_rate=0.01, teta_tolerance=0.0001, iteration_tolerance=10000)
```

```python

xs = np.array([np.min(x), np.max(x)])
ys = teta_zero + teta_one*xs

fig, ax = plt.subplots()
plt.title("Train Data & Regression")
ax.scatter(x, y, label="Train Data")
ax.plot(xs,ys, "r",label="Regression")
ax.legend()

plt.show()
```
